```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler

```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
spaceship.shape
```

**Check for data types**

```python
spaceship.dtypes
```

**Check for missing values**

```python
spaceship.isnull().sum()
```

```python
spaceship.isna().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
ss_cleaned = spaceship.dropna()
```

```python
print("Original data shape:", spaceship.shape)
print("Cleaned data shape:", ss_cleaned.shape)
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
ss_cleaned['Cabin'].unique()
```

```python
ss_cleaned.loc[:, 'Cabin'] = ss_cleaned['Cabin'].str[0]
```

```python
print(ss_cleaned['Cabin'].unique())
```

- Drop PassengerId and Name

```python
ss_cleaned = ss_cleaned.drop(['PassengerId', 'Name'], axis=1)
```

```python
print(ss_cleaned.columns)
```

- For non-numerical columns, do dummies.

```python
non_numerical_cols = ss_cleaned.select_dtypes(include=['object']).columns
print(non_numerical_cols)
```

```python
ss_cleaned = pd.get_dummies(ss_cleaned, columns=non_numerical_cols, drop_first=True)
```

```python
print(ss_cleaned.head())
print(ss_cleaned.columns)
```

**Perform Train Test Split**

```python
X = ss_cleaned.drop('Transported', axis=1)  # Replace with the name of your target column
y = ss_cleaned['Transported']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

- Evaluate your model's performance. Comment it

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)  # Adjust 'k' as needed
knn.fit(X_train_scaled, y_train)
```

```python
y_pred_scaled = knn.predict(X_test_scaled)
```

```python
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
print("Scaled Accuracy:", accuracy_scaled)
```

```python
cm_scaled = confusion_matrix(y_test, y_pred_scaled)
print("Confusion Matrix (Scaled):\n", cm_scaled)
```

```python
report_scaled = classification_report(y_test, y_pred_scaled)
print("Classification Report (Scaled):\n", report_scaled)
```
